```python
# Import data file
worldCoord = "dataUsed/worldcities.csv"

# Read data file, need the encoding to read properly
worldCoord_raw = pd.read_csv(worldCoord, encoding="ISO-8859-1")
display(worldCoord_raw.head(2))
display(worldCoord_raw.tail(2))
```

```python
filter_worldCoord = worldCoord_raw[worldCoord_raw["capital"]=="primary"]
filter_worldCoord.reset_index(drop=True, inplace=True)
filter_worldCoord
```

```python
list(filter_worldCoord.columns)
```

```python
secondfilter_worldCoord = filter_worldCoord[['city','lat','lng','country','iso3','population']]
secondfilter_worldCoord
```

```python
# Rename the ISO_A3 to iso_code for merging
worldCoor_renamed = secondfilter_worldCoord.rename(columns={"iso3":"iso_code", "country":"location"})
worldCoor_renamed
```

```python
# Merge all
completeCoor = pd.merge(world_raw, worldCoor_renamed, on=["iso_code", "location"], how="left")
display(completeCoor.head(2))
display(completeCoor.tail(2))
```

```python
# geoMergeCol = list(geoMerge.columns)
# geoMergeCol
```

```python
completeCoorCol = list(completeCoor.columns)
completeCoorCol
```

```python
def df_to_geojson(df, properties, lat='lat', lon='lng'):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson
```
